```python
## Stack as List

stack=[]

stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)


print(stack)

#top element
print(f"top element: {stack[-1]}")

## getting the top element
print(stack.pop())

#top element
print(f"top element: {stack[-1]}")

print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())

```

```python
## Queue as List

queue=[]

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)

print(queue)

 #remove again
print(queue.pop(0))

#access

print(f"front e ache {queue[0]}")

print(queue.pop(0))

#access

print(f"front e ache {queue[0]}")


#top element
```

### List Comprehension

```python
even=[]


#navie way
for i in range(100):
    if i%2==0:
        even.append(i)
print(even)
```

```python
random_list=[x*2 for x in range(10) if x%3==0]

print(random_list)
```

```python
fruits=['apple','banana','orange']

capitalize_fruits=[]

for fruit in fruits:
    capitalize_fruits.append(fruit.capitalize())
print(capitalize_fruits)
```

```python
fruits=['apple','banana','orange']

lower_fruits=[fruit.lower() for fruit in fruits]

print(lower_fruits)
```
