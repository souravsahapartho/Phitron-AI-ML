## String Methods


```python
string="Welcome to Phitron ML Course, Phitron, Learn ML, read ML"

processed_string= string.lower()

print(string)

print(processed_string)

processed_upper=string.upper()
print(processed_upper)

processed_capitalize=string.capitalize()
print(processed_capitalize)

#find length
print(len(string))

print("Phitron" in string) ##find word


#suru r kon index theke substring ta suru hoyeche
shurur_index =processed_string.find("phitron")
print(shurur_index)

#sesh r kon index theke substring ta suru hoyeche
seshr_index=processed_string.rfind("phitron")
print(seshr_index)


##substring koybar ache, seta ber korbo
count=processed_string.count("phitron")
print(count)

## word k onno word diye replace korbo
new_string=string.replace("ML", "AI,ML")
print(new_string)




```

## Spliting, Joining, Formatted STring

```python
prompt="What is python ?"

tokens=prompt.split()

print(tokens)

sentence= "-".join(tokens)

print(sentence)
```
