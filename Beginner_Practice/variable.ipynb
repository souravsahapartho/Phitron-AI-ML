## Max and Min Problem


```python
inp=input()

numbers=inp.split()

x=int(numbers[0])
y=int(numbers[1])
z=int(numbers[2])

min=x
max=x

## minimum number
if min>y:
    min=y
if min>z:
    min=z

##maximum number

if y>max:
    max=y
if z>max:
    max=z

print(min, max)
```

#### Even, Odd, Positive and Negative


```python
n=int(input())

inp=input()

numbers=inp.split()

positive=0; negative=0; even=0; odd=0;

for i in range(n):
    x=int(numbers[i])

    if x>0 :
        positive+=1
    elif x<0:
        negative+=1
    
    #odd or even
    if x%2!=0:
        odd+=1

    else:
        even+=1
    
print("Even: ",even)
print("Odd: ",odd)
print("Positive: ",positive)
print("Negative: ",negative)
```

### Digits Reverse
